## Labels

1. select dataset
2. enter from waiting state
3. append text
4. modify text --> update text message to data
5. modify y-axis to above 2px from data --> eg: height = 300 and data 100, so (300 - 100 - 2 = 198), label will be on 198 (from above) and (300 - 198 = 102) from below
6. modify x-axis to be the same position as the data --> eg: barwidth 55.56 and i = 1 (index[1]), so (55.56 * 1)
7. modify label color
